##Mtodo Orientado a objetos

```python
fig= plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
```

```python
fig= plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])
axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')
fig.show()
```

##Subplots

```python
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2)
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)
ax5.plot(x,1/np.cos(x),'bo-')
ax6.plot(x,1/np.sin(x),'yD--')
ax7.plot(x,1/np.tan(x),'rs-')
ax8.plot(x,1/np.cos(x)**2,'gD--')
fig.tight_layout()
```

##Contexto para grficas

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,y, label='$sin(x)$')
axes[0].set_title('Relacin X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y,x)
axes[1].set_title('Relacin Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x,y,label='$sin(x)$')
plt.title('Relacin X - Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left', bbox_to_anchor=(1,0.9))
```

```python
plt.plot(x,y,label='$sin(x)$')
plt.title('Relacin X - Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

##Colores y Estilos

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2

print(plt.style.available)
```

```python
for i in plt.style.available:
  plt.style.use(i)
  fig, ax = plt.subplots(figsize=(2,2))
  ax.plot(x, y)
  ax.set_title(i, size=20)
```

```python
plt.style.use('dark_background')
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(x,x+1,'r')
ax.plot(x,x+2,'b')
ax.plot(x,x+3,'y')
ax.plot(x,x+4,'g')
```

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

##Pyplot

```python
fig, ax = plt.subplots(figsize=(5,5))

ax.plot(x,x+1,color='green', alpha=0.8, linewidth=5, linestyle='-', marker='o', markersize=8,
        markerfacecolor='#CCFFCC')
ax.plot(x,x+2,color='blue', linewidth=8, linestyle='--', marker='8')
ax.plot(x,x+3,color='#66FF66', linewidth=3, linestyle='dashed', marker='v')
ax.plot(x,x+4,color='#CCFFCC', linewidth=2, linestyle=':', marker='P')
```
