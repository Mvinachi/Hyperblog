```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
plt.plot(x,y,'rs-')
plt.show()
```

```python
plt.plot(x,y,'yD:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

##Subplot

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```

##Método Orientado a objetos

```python
fig= plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
```

```python
fig= plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])
axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')
fig.show()
```

##Subplots

```python
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,np.cos(x)**2)
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'y')
ax3.plot(x,np.tan(x),'r')
ax4.plot(x,np.cos(x)**2)
ax5.plot(x,1/np.cos(x),'bo-')
ax6.plot(x,1/np.sin(x),'yD--')
ax7.plot(x,1/np.tan(x),'rs-')
ax8.plot(x,1/np.cos(x)**2,'gD--')
fig.tight_layout()
```
